### Отчет по Упражнению 4

**Задание:** Найти значение параметра `a`, при котором функция $f(x)$ является непрерывной в точке $x=0$, где $f(x)$ определена кусочно:
$$ f(x) = \begin{cases} x \ln(x^2), & x > 0 \\ a, & x = 0 \\ \frac{\sinh(x)}{x} - 1, & x < 0 \end{cases} $$

**Метод:** Для того чтобы функция была непрерывной в точке $x=0$, должны выполняться следующие условия:
1.  Существует правый предел $\lim_{x \to 0^+} f(x)$.
2.  Существует левый предел $\lim_{x \to 0^-} f(x)$.
3.  Значение функции в точке $x=0$ равно этим пределам: $f(0) = \lim_{x \to 0^+} f(x) = \lim_{x \to 0^-} f(x)$.

Использована библиотека `sympy.limit` для вычисления односторонних пределов и `sympy.solve` для решения уравнения относительно `a`.

**Результаты:**
*   **Правый предел ($x \to 0^+$):** $\lim_{x \to 0^+} x \ln(x^2) = 0$
*   **Левый предел ($x \to 0^-$):** $\lim_{x \to 0^-} (\frac{\sinh(x)}{x} - 1) = \lim_{x \to 0^-} (\frac{x + x^3/6 + ...}{x} - 1) = \lim_{x \to 0^-} (1 + x^2/6 + ... - 1) = 0$
*   **Значение `a`:** Для непрерывности функции в $x=0$, $a$ должно быть равно значению односторонних пределов. Таким образом, $a = 0$.

**Вывод:** Функция $f(x)$ будет непрерывной в точке $x=0$, если $a=0$.

### Отчет по Упражнению 4

**Задание:** Найти значение параметра `a`, при котором функция $f(x)$ является непрерывной в точке $x=0$, где $f(x)$ определена кусочно:
$$ f(x) = \begin{cases} x \ln(x^2), & x > 0 \\ a, & x = 0 \\ \frac{\sinh(x)}{x} - 1, & x < 0 \end{cases} $$

**Метод:** Для того чтобы функция была непрерывной в точке $x=0$, должны выполняться следующие условия:
1.  Существует правый предел $\lim_{x \to 0^+} f(x)$.
2.  Существует левый предел $\lim_{x \to 0^-} f(x)$.
3.  Значение функции в точке $x=0$ равно этим пределам: $f(0) = \lim_{x \to 0^+} f(x) = \lim_{x \to 0^-} f(x)$.

Использована библиотека `sympy.limit` для вычисления односторонних пределов и `sympy.solve` для решения уравнения относительно `a`.

**Результаты:**
*   **Правый предел ($x \to 0^+$):** $\lim_{x \to 0^+} x \ln(x^2) = 0$
*   **Левый предел ($x \to 0^-$):** $\lim_{x \to 0^-} (\frac{\sinh(x)}{x} - 1) = \lim_{x \to 0^-} (\frac{x + x^3/6 + ...}{x} - 1) = \lim_{x \to 0^-} (1 + x^2/6 + ... - 1) = 0$
*   **Значение `a`:** Для непрерывности функции в $x=0$, $a$ должно быть равно значению односторонних пределов. Таким образом, $a = 0$.

**Вывод:** Функция $f(x)$ будет непрерывной в точке $x=0$, если $a=0$.
