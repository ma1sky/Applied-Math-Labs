```python
#a
def M1(f, g, a, b, n):
  x = np.linspace(a, b, n)
  plt.plot(x, f(x))
  plt.plot(x, g(x))
  plt.fill_between(x, f(x), g(x), where=(f(x) <= g(x)), color='blue', alpha=0.4, hatch='|')
  plt.plot(x, np.maximum(f(x), g(x)), linewidth=3, color='blue')
  plt.plot(x, np.minimum(f(x), g(x)), linewidth=3, color='blue')
  plt.grid(True)
  plt.show()
#б
def M2(f, g, a, b, n):
  x = np.linspace(a, b, n)
  plt.plot(x, f(x))
  plt.plot(x, g(x))
  plt.fill_between(x, f(x), g(x), where=(f(x) <= g(x)), color='blue', alpha=0.4, hatch='|')
  plt.plot(x, np.maximum(f(x), g(x)), linewidth=3, color='blue', linestyle="dashed")
  plt.plot(x, np.minimum(f(x), g(x)), linewidth=3, color='blue', linestyle="dashed")
  plt.grid(True)
  plt.show()
#в
def M3(f, g, a, b, n):
  x = np.linspace(a, b, n)
  plt.plot(x, f(x))
  plt.plot(x, g(x))
  plt.fill_between(x, f(x), g(x), where=(f(x) <= g(x)), color='blue', alpha=0.4, hatch='|')
  plt.plot(x, np.maximum(f(x), g(x)), linewidth=3, color='blue', linestyle="dashed")
  plt.plot(x, np.minimum(f(x), g(x)), linewidth=3, color='blue')
  plt.grid(True)
  plt.show()
#г
def M4(f, g, a, b, n):
  x = np.linspace(a, b, n)
  plt.plot(x, f(x))
  plt.plot(x, g(x))
  plt.fill_between(x, f(x), g(x), where=(f(x) <= g(x)), color='blue', alpha=0.4, hatch='|')
  plt.plot(x, np.minimum(f(x), g(x)), linewidth=3, color='blue', linestyle="dashed")
  plt.plot(x, np.maximum(f(x), g(x)), linewidth=3, color='blue')
  plt.grid(True)
  plt.show()

a, b, n = -4, 3, 100
f = lambda x: x**2+4*x+1
g = lambda x: np.cos(x)

M4(f, g, a, b, n)
```
