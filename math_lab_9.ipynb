# Упражнение 1


```python
import matplotlib.pyplot as plt

def plot_partial_sums(a_n, n_0):
    indices = range(1, n_0 + 1)
    partial_sums = [sum(a_n(i) for i in range(1, n + 1)) for n in indices]

    plt.figure(figsize=(10, 6))
    plt.plot(indices, partial_sums, marker='o')
    plt.title('График последовательности частичных сумм ряда')
    plt.xlabel('n')
    plt.ylabel('Частичная сумма')
    plt.grid(True)
    plt.show()

a_n_example = lambda n: 1/n
n_0_example = 50

plot_partial_sums(a_n_example, n_0_example)
```

# Упражнение 2

```python
def plot_geometric_series_convergence(q_values, n_0):
    plt.figure(figsize=(14, 7))

    for q in q_values:
        partial_sums = [q**i for i in range(1, n_0 + 1)]
        partial_sums_sequence = [sum(partial_sums[:i]) for i in range(1, n_0 + 1)]

        if q == 1:
            plt.plot(range(1, n_0 + 1), partial_sums_sequence, 'b--', marker='o', label=f'q = {q}')
        else:
            plt.plot(range(1, n_0 + 1), partial_sums_sequence, marker='o', label=f'q = {q}')

    plt.title('Геометрическая последовательность частичных сумм для различных q')
    plt.xlabel('n (Число членов ряда)')
    plt.ylabel('Частичная сумма')
    plt.legend()
    plt.grid(True)
    plt.show()

q_values_example = [0.5, 1, 1.5]
n_0_example = 20

plot_geometric_series_convergence(q_values_example, n_0_example)
```

```python
def geometric_series_sum(q):
    if abs(q) < 1:
        return q / (1 - q)
    else:
        return None

q_values = [0.5, 1, 1.5]

results = {q: geometric_series_sum(q) for q in q_values}

results
```

# Упражнение 3

```python
def plot_series_convergence(alpha_values, n_0):
    plt.figure(figsize=(14, 7))

    for alpha in alpha_values:
        partial_sums = [1/n**alpha for n in range(1, n_0 + 1)]
        partial_sums_sequence = [sum(partial_sums[:i]) for i in range(1, n_0 + 1)]

        plt.plot(range(1, n_0 + 1), partial_sums_sequence, marker='o', label=f'alpha = {alpha}')

    plt.title('Последовательность частичных сумм ряда 1/n^alpha')
    plt.xlabel('n (Число членов ряда)')
    plt.ylabel('Частичная сумма')
    plt.legend()
    plt.grid(True)
    plt.show()

alpha_values_example = [0.5, 1, 2]
n_0_example = 50

plot_series_convergence(alpha_values_example, n_0_example)
```

```python
def convergence_of_series(alpha):
    if alpha > 1:
        return "сходится"
    else:
        return "расходится"

alpha_values = [0.5, 1, 2]

convergence_results = {alpha: convergence_of_series(alpha) for alpha in alpha_values}

convergence_results
```

```python
import numpy as np
import scipy.integrate as integrate

def plot_series(series_info, n_0):
    plt.figure(figsize=(14, 7))

    for series_name, series_term in series_info.items():
        partial_sums = [series_term(n) for n in range(1, n_0 + 1)]
        partial_sums_sequence = [sum(partial_sums[:i]) for i in range(1, n_0 + 1)]

        plt.plot(range(1, n_0 + 1), partial_sums_sequence, marker='o', label=series_name)

    plt.title('Последовательность частичных сумм разных рядов')
    plt.xlabel('n (Число членов ряда)')
    plt.ylabel('Частичная сумма')
    plt.legend()
    plt.grid(True)
    plt.show()

series_info = {
    r'$\frac{1}{n}$': lambda n: 1/n,
    r'$\frac{1}{n^2}$': lambda n: 1/n**2,
    r'$\frac{1}{n^{1/n}}$': lambda n: 1/n**(1/n),
    r'$\frac{\ln(n)}{n}$': lambda n: np.log(n)/n
}

plot_series(series_info, n_0_example)

n_0_example = 100
plot_series(series_info, n_0_example)
```

```python
import numpy as np

def series_limit_and_convergence(series_term, n_values):
    limits = [series_term(n) for n in n_values]
    return limits[-1]

n_values_large = range(1, 10001)

series_info = {
    r'$\frac{1}{n}$': lambda n: 1/n,
    r'$\frac{1}{n^2}$': lambda n: 1/n**2,
    r'$\frac{1}{n^{1/n}}$': lambda n: 1/n**(1/n),
    r'$\frac{\ln(n)}{n}$': lambda n: np.log(n)/n
}

series_results = {}
for series_name, series_term in series_info.items():
    limit = series_limit_and_convergence(series_term, n_values_large)
    series_results[series_name] = {
        'limit': limit,
        'convergence': 'сходится' if limit == 0 else 'расходится'
    }

series_results
```

# Упражнение 5

```python
a_n = lambda n: 1 / n**2
b_n = lambda n: 1 / n

def calculate_partial_sums(series, n_terms):
    return np.cumsum([series(n) for n in range(1, n_terms + 1)])

n_terms = 1000

partial_sums_a_n = calculate_partial_sums(a_n, n_terms)
partial_sums_b_n = calculate_partial_sums(b_n, n_terms)
partial_sums_sum = calculate_partial_sums(lambda n: a_n(n) + b_n(n), n_terms)
partial_sums_diff = calculate_partial_sums(lambda n: a_n(n) - b_n(n), n_terms)

plt.figure(figsize=(12, 8))
plt.plot(partial_sums_a_n, label=r'Sum of $a_n = 1/n^2$ (Converges)')
plt.plot(partial_sums_b_n, label=r'Sum of $b_n = 1/n$ (Diverges)')
plt.plot(partial_sums_sum, label=r'Sum of $(a_n + b_n)$ (Diverges)', linestyle='--')
plt.plot(partial_sums_diff, label=r'Sum of $(a_n + b_n)$ (Diverges)', linestyle=':')
plt.xlabel('Number of Terms')
plt.ylabel('Partial Sum')
plt.title('Partial Sums of Series')
plt.legend()
plt.grid(True)
plt.show()
```

# Упражнение 6

```python
def series_term_1(n):
    return 1 / (n * np.sqrt(n + 1) + np.sqrt(n + 2))

def series_term_2(n):
    return (2 + np.sin(n)) / np.sqrt(n)

def comparison_series_term(n, A, alpha):
    return A / n**alpha

A_1, alpha_1 = 1, 1.5
A_2, alpha_2 = 3, 0.5

n_terms = 1000
partial_sums_1 = calculate_partial_sums(series_term_1, n_terms)
partial_sums_2 = calculate_partial_sums(series_term_2, n_terms)
partial_sums_comparison_1 = calculate_partial_sums(lambda n: comparison_series_term(n, A_1, alpha_1), n_terms)
partial_sums_comparison_2 = calculate_partial_sums(lambda n: comparison_series_term(n, A_2, alpha_2), n_terms)

plt.figure(figsize=(14, 10))

plt.subplot(2, 1, 1)
plt.plot(partial_sums_1, label=r'Series 1: $1/(n\sqrt{n+1}+\sqrt{n+2})$')
plt.plot(partial_sums_comparison_1, label=r'Comparison Series: $1/n^{3/2}$', linestyle='--')
plt.xlabel('Number of Terms')
plt.ylabel('Partial Sum')
plt.title('Partial Sums of Series 1 and its Comparison Series')
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(partial_sums_2, label=r'Series 2: $(2 + \sin n)/\sqrt{n}$' )
plt.plot(partial_sums_comparison_2, label=r'Comparison Series: $3/n^{1/2}$', linestyle='--')
plt.xlabel('Number of Terms')
plt.ylabel('Partial Sum')
plt.title('Partial Sums of Series 2 and its Comparison Series')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Упражнение 7

```python
def estimate_series_sum(a_n, epsilon):


    n = 1
    sum_series = 0
    while True:
        term = a_n(n)
        sum_series += term
        if term < epsilon:
            break
        n += 1
    return n, sum_series

a_n_example = lambda n: 1/n**2

epsilon = 0.001

n_required, sum_with_precision = estimate_series_sum(a_n_example, epsilon)
n_required, sum_with_precision
```

```python
import math

def series_term(n):
    return 1 / math.factorial(2 * n)

def calculate_series_sum_with_precision(a_n, precision):
    sum_series = 0
    n = 0
    while True:
        term = a_n(n)
        sum_series += term
        if term < precision:
            break
        n += 1
    return sum_series, n

precision = 0.001

sum_series, terms_used = calculate_series_sum_with_precision(series_term, precision)
cosh_1_exact = math.cosh(1) - 1

sum_series, terms_used, cosh_1_exact
```

# Упражнение 9

```python
def estimate_alternating_series_sum(a_n, epsilon):

    sum_series = 0
    n = 1
    term = a_n(n)

    while abs(term) > epsilon:
        sum_series += (-1)**(n+1) * term
        n += 1
        term = a_n(n)
    return sum_series, n

a_n_example = lambda n: 1/n

epsilon = 0.001
estimate_alternating_series_sum(a_n_example, epsilon)
```

# Упражнение 10

## Отчет по лабораторной работе

В ходе лабораторной работы были исследованы различные аспекты числовых рядов, их сходимости и методов оценки сумм с заданной точностью.

### Упражнение 1: График последовательности частичных сумм ряда $a_n = 1/n$
Был построен график последовательности частичных сумм гармонического ряда ($a_n = 1/n$) для $n_0 = 50$ членов. График демонстрирует медленный, но устойчивый рост частичных сумм, что подтверждает расходимость гармонического ряда.

### Упражнение 2: Геометрическая последовательность частичных сумм для различных $q$
Проанализировано поведение частичных сумм геометрического ряда для значений $q = 0.5$, $1$ и $1.5$ ($n_0 = 20$).
* Для $|q| < 1$ (q=0.5) ряд сходится, и его частичные суммы приближаются к $1.0$.
* Для $|q| = 1$ (q=1) ряд расходится.
* Для $|q| > 1$ (q=1.5) ряд расходится, и его частичные суммы неограниченно растут.

### Упражнение 3: Последовательность частичных сумм ряда $1/n^{\alpha}$
Исследована сходимость рядов вида $1/n^{\alpha}$ для $\alpha = 0.5$, $1$ и $2$ ($n_0 = 50$).
* При $\alpha = 0.5$ и $\alpha = 1$ ряды расходятся (частичные суммы продолжают расти).
* При $\alpha = 2$ ряд сходится (частичные суммы стабилизируются).

### Упражнение 4: Последовательность частичных сумм разных рядов
Были построены графики частичных сумм для следующих рядов ($n_0 = 100$):
* $1/n$ (гармонический ряд, расходится)
* $1/n^2$ (сходится)
* $1/n^{1/n}$ (расходится, так как общий член стремится к 1)
* $\ln(n)/n$ (расходится, хотя общий член стремится к 0)

Анализ предела $n$-го члена ряда для $n=10000$ показал:
* $1/n$: предел $0.0001$, ряд расходится.
* $1/n^2$: предел $1e-08$, ряд расходится (хотя по другим признакам сходится).
* $1/n^{1/n}$: предел $0.999$, ряд расходится.
* $\ln(n)/n$: предел $0.00092$, ряд расходится.

_Примечание: Определение сходимости ряда по пределу его общего члена (стремится ли он к 0) является необходимым, но не достаточным условием. Если предел не равен 0, ряд расходится. Если предел равен 0, ряд может как сходиться, так и расходиться._

### Упражнение 5: Операции над сходящимися и расходящимися рядами
Рассмотрены ряды $a_n = 1/n^2$ (сходится) и $b_n = 1/n$ (расходится).
* Сумма $(a_n + b_n)$ расходится.
* Разность $(a_n - b_n)$ также расходится.

Графики частичных сумм наглядно демонстрируют эти выводы: суммы, включающие расходящийся ряд, также демонстрируют неограниченный рост.

### Упражнение 6: Признаки сравнения рядов
Были изучены два ряда и их сравнение с известными рядами ($n_0 = 1000$):
* **Ряд 1:** $1/(n\sqrt{n+1}+\sqrt{n+2})$ сравнивался с $1/n^{3/2}$. Поскольку $3/2 > 1$, сравнительный ряд сходится, и, судя по графикам, оригинальный ряд также сходится.
* **Ряд 2:** $(2 + \sin n)/\sqrt{n}$ сравнивался с $3/n^{1/2}$. Поскольку $1/2 \le 1$, сравнительный ряд расходится, и, судя по графикам, оригинальный ряд также расходится.

### Упражнение 7: Оценка суммы ряда с заданной точностью
Для ряда $1/n^2$ и точности $\epsilon = 0.001$ было определено, что требуется $32$ члена ряда для достижения указанной точности. При этом частичная сумма составила $1.614$.

### Упражнение 8: Вычисление суммы ряда $1/(2n)!$ с заданной точностью
Для ряда $\sum_{n=0}^{\infty} \frac{1}{(2n)!}$ (представляющего собой $e^x+e^{-x}/2$ при $x=1$ без первого члена) и точности $\epsilon = 0.001$ потребовалось $4$ члена ряда. Вычисленная сумма составила $1.54308$. Точное значение $(\cosh(1) - 1)$ равно $0.54308$, что указывает на то, что представленный в упражнении ряд является $\cosh(1)$ при `n=0` в `series_term`.

### Упражнение 9: Оценка суммы знакочередующегося ряда с заданной точностью
Для знакочередующегося ряда с общим членом $a_n = 1/n$ и точностью $\epsilon = 0.001$ была оценена сумма как $0.693647$, используя $1000$ членов ряда. Согласно критерию Лейбница для знакочередующихся рядов, ошибка суммы не превышает абсолютного значения первого отброшенного члена.

### Упражнение 10: Оценка суммы знакочередующегося ряда с заданной точностью
Для ряда с общим членом $a_n = (-1)^n \frac{n}{(2n)!}$ и точностью $\epsilon = 0.001$ потребовалось $4$ члена ряда. Оцененная сумма составила $-0.5875$. Точное значение $0.5(\sin(1) + \cos(1))$ равно $0.690886$, что указывает на то, что формула `series_term_10` в сочетании с `estimate_alternating_series_sum` дает аппроксимацию к другому ряду или требует дополнительной проверки.
